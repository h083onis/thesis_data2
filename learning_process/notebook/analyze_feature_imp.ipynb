```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../resource/feature_importance2.csv', index_col=0, header=0)
df
```

```python
df['project'] = df.index.map(lambda x: x.split('-')[0])
df
```

```python
df2 = df.groupby('project').mean()
df2
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig, axes = plt.subplots(2, 1,  figsize=(16, 8), tight_layout=True)

for i, row in enumerate(df2.iterrows()):
    axes[i].bar(df2.columns, row[1], label=row[0])
    
# プロットの表示
for i in range(2):
    axes[i].set_xlabel('columns')
    axes[i].set_ylabel('value')
axes[0].set_title("OpenStack feature importance") 
axes[1].set_title("Qt feature importance")
plt.show()
```
